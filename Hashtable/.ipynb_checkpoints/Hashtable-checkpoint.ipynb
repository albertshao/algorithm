```python
# initial capacity for internal array 
INITIAL_CAPACITY = 50 

# Node data structure -- essential a LinkedList node
class Node(object):
    def __init__(self, key, value):
        self.key = key 
        self.value = value 
        self.next = None 
        
    def __str__(self):
        return '<Node(%s:%s), next:%s>' % (self.key, self.value, self.next != None)
    
    def __repr__(self):
        return str(self)
    

# Hash table with separate chaining 
class HashTable:
    def __init__(self):
        """
        Initialize hash table
        """
        self.capacity = INITIAL_CAPACITY
        
        # the real number of node. 
        self.size = 0 
        self.buckets = [None] * self.capacity
    
    def hash(self, key):
        """
        Generate a hash for a given key 
        Args: 
            key: a string, a given key 
        Returns:
            hashsum: a integer, value from 0 to self.capacity 
        """
        hashsum = 0 
        for idx, c in enumerate(key):
            # ord() method returns an integer representing the unicode code point of given unicode character 
            hashsum += (idx + len(key)) ** ord(c)  
            hashsum = hashsum % self.capacity 
        return hashsum
    
    def insert(self, key, value):
        """
        Insert the a key, value pair to the hashtable 
        Args:
            key: a string, the key to value 
            value: anything type of object
        """
        # 1. Increment size 
        self.size += 1 
        # 2. Compute index of key 
        index = self.hash(key)
        # Go to the node corresponding to the hash 
        node = self.buckets[index]
        # 3. If the bucket is empty 
        if node is None:
            # Create node, add it, return 
            self.buckets[index] = Node(key, value)
            return 
        
        # 4. Iterate to the end of the linked list at the provied index 
        prev = node 
        while node is not None:
            prev = node 
            node = node.next 
        
        # Add a new node at the end of the list with provided key/value 
        prev.next = Node(key, value)
        
        
    def find(self, key):
        """
        Find a data vlaue based on key 
        Args:
            key: a string, the given key 
        Returns:
            value: any type, value stored under key or None if not found. 
        """
        # 1. Compute the hash 
        index = self.hash(key)
        # 2. Go to first node in list at bucket 
        node = self.buckets[index]
        # 3. Traverse the linked list at this node 
        while node is not None and node.key != key:
            node = node.next 
        # 4. Now, node is the requested key/value pair or None 
        if node is None:
            # not found 
            return None 
        else:
            # Found -return the data value 
            return node.value 
        
    def remove(self, key):
        """
        Remove the node stored under key 
        Args:
            key: a string, the specific key 
        Returns:
            removed data value or None if not found 
        """
        # 1. Compute hash 
        index = self.hash(key)
        node = self.buckets[index]
        prev = None 
        
        # Iterate to the quested node 
        while node is not None and node.key != key:
            prev = node
            node = node.next
        # Now, node is either the quested node or none
        if node is None:
            # 3. key not found 
            return None 
        else: 
            # 4. The key was found 
            self.size -= 1
            result = node.value 
            
            # Delete this element in linked list 
            if prev is None: 
                self.buckets[index] = node.next
            else:
                # LinkedList delete by skipping over 
                prev.next = prev.next.next  
            
            # Return the delete result 
            
            return result 
        
        
        
```

```python
import unittest

class TestHashTable(unittest.TestCase):
    def setUp(self):
        self.ht = HashTable()
    def test_hash(self):
        self.assertEqual(self.ht.hash("hello"), self.ht.hash("hello"))
        self.assertTrue(self.ht.hash("hello") < self.ht.capacity)
    def test_insert(self):
        self.assertEqual(self.ht.size, 0)
        self.ht.insert("test_key", "test_value")
        self.assertEqual(self.ht.size, 1)
        self.assertEqual(self.ht.buckets[self.ht.hash("test_key")].value, "test_value")
    def test_find(self):
        self.assertEqual(self.ht.size, 0)
        obj = "hello"
        self.ht.insert("key1", obj)
        self.assertEqual(obj, self.ht.find("key1"))
        obj = ["this", "is", "a", "list"]
        self.ht.insert("key2", obj)
        self.assertEqual(obj, self.ht.find("key2"))
    def test_remove(self):
        self.assertEqual(self.ht.size, 0)
        obj = "test object"
        self.ht.insert("key1", obj)
        self.assertEqual(1, self.ht.size)
        self.assertEqual(obj, self.ht.remove("key1"))
        self.assertEqual(0, self.ht.size)
        self.assertEqual(None, self.ht.remove("some random key"))
    def test_capacity(self):
        # Test all public methods in one run at a large capacity
        for i in range(0,1000):
            self.assertEqual(i, self.ht.size)
            self.ht.insert("key" + str(i), "value")
        self.assertEqual(self.ht.size, 1000)
        for i in range(0,1000):
            self.assertEqual(1000-i, self.ht.size)
            self.assertEqual(self.ht.find("key" + str(i)), self.ht.remove("key" + str(i)))
    def test_issue2(self):
        self.assertEqual(self.ht.size, 0)
        self.ht.insert('A', 5)
        self.assertEqual(self.ht.size, 1)
        self.ht.insert('B', 10)
        self.assertEqual(self.ht.size, 2)
        self.ht.insert('Ball', 'hello')
        self.assertEqual(self.ht.size, 3)

        self.assertEqual(5, self.ht.remove('A'))
        self.assertEqual(self.ht.size, 2)
        self.assertEqual(None, self.ht.remove('A'))
        self.assertEqual(self.ht.size, 2)
        self.assertEqual(None, self.ht.remove('A'))
        self.assertEqual(self.ht.size, 2)
```

```python
tht = TestHashTable()
tht.setUp()
tht.test_hash()
tht.test_insert()
#tht.test_find()
#tht.test_remove()
```
